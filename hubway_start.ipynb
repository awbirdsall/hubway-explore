## use lxml

```python
hubwaylxml = etree.parse(urllib.urlopen('http://www.thehubway.com/data/stations/bikeStations.xml'))
```

```python
hubwaylxml.findall('station')
```

```python
# last comm with server is microseconds since 12/31/1969. Seems to account for daylight savings
for elt in hubwaylxml.getiterator(tag='lastCommWithServer'):
    print datetime.datetime.fromtimestamp(int(elt.text)/1.e3)
```

```python
for elt in hubwaylxml.getiterator():
    print elt
```

```python
sorted(tname.text for tname in hubwaylxml.getiterator('terminalName'))
```
